Importing pandas and numpy


```python
import pandas as pd
import numpy as np
```

Reading the csv file which i extracted from the website


Extracted the details of all three companies from the website and put in a excel file and then converted into a csv file for further usage.


```python
df=pd.read_csv('Financials_2021-2023.csv')
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

Finding revenue growth and net income growth

```python
# Fill missing values in the 'Company' column with the previous valid value
df['Total Revenue'] = df['Total Revenue'].replace({'\$': '', ',': '', 'B': 'e9', 'M': 'e6'}, regex=True).astype(float)
df['Net Income'] = df['Net Income'].replace({'\$': '', ',': '', 'B': 'e9', 'M':
                                             'e6'}, regex=True).astype(float)
df['Company'] = df['Company'].fillna(method='ffill')

df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100

# Display the updated DataFrame to verify the changes
df
```

Aggregate values:

```python
# Aggregation by company
company_summary = df.groupby('Company').agg({
    'Total Revenue': ['mean', 'sum', 'max'],
    'Net Income': ['mean', 'sum', 'max'],
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()

# Aggregation by fiscal year


# Display summaries
print("Company Summary:")
print(company_summary)
```

```python
def simple_chatbot(user_query):
    # Example predefined queries
    if user_query == "What is the total revenue?":
        # Example: Return total revenue for all companies in 2023
        result = ""
        for company in df['Company'].unique():
            rev = df[(df['Company'] == company) & (df['Fiscal Year'] == 2023)]['Total Revenue'].values[0]
            result += f"{company} total revenue (2023): ${rev}B\n"
        return result.strip()

    elif user_query == "How has net income changed over the last year?":
        result = ""
        for company in df['Company'].unique():
            ni_2023 = df[(df['Company'] == company) & (df['Fiscal Year'] == 2023)]['Net Income'].values[0]
            ni_2022 = df[(df['Company'] == company) & (df['Fiscal Year'] == 2022)]['Net Income'].values[0]
            diff = ni_2023 - ni_2022
            direction = "increased" if diff > 0 else "decreased"
            result += f"{company} net income has {direction} by ${abs(diff):.2f}B from 2022 to 2023\n"
        return result.strip()

    elif user_query == "What is the profit margin?":
        result = ""
        for company in df['Company'].unique():
            rev = df[(df['Company'] == company) & (df['Fiscal Year'] == 2023)]['Total Revenue'].values[0]
            ni = df[(df['Company'] == company) & (df['Fiscal Year'] == 2023)]['Net Income'].values[0]
            margin = (ni / rev) * 100
            result += f"{company} profit margin (2023): {margin:.2f}%\n"
        return result.strip()

    else:
        return "Sorry, I can only provide information on predefined queries."
```

```python
# Example test runs
print(simple_chatbot("What is the total revenue?"))
print(simple_chatbot("How has net income changed over the last year?"))
print(simple_chatbot("What is the profit margin?"))
print(simple_chatbot("Who is the CEO?"))
```

```python
while True:
    user_query = input("Ask a financial question (or type 'exit' to quit): ")
    if user_query.lower() == 'exit':
        break
    print(simple_chatbot(user_query))
```
